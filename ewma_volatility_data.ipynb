```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/NSQ_OneYear100test_Sept21.csv")
```

```python
# read csv file
df
```

```python
# divide the Local_Date_Time column into Date and Time columns
df = df.iloc[:, 1:]
df['Date'] = pd.to_datetime(df['Local_Date_Time'], format='%Y-%m-%d').dt.date
df['Time'] = pd.to_datetime(df['Local_Date_Time'], format='%Y-%m-%d').dt.time

df = df[['BarTime', 'Date', 'Time', 'RIC', 'Open', 'High', 'Low', 
         'Close','VWAP', 'NumberOfTrades', 'Volume']]
```

```python
df
```

```python
df_RIC = df.groupby(["RIC"])
df_RIC.count()
```

```python
# list of companies in an array
l_company = df["RIC"].unique()
```

```python
# for someone not too familiar with python
def find_company_index(company_name):
    index = np.where(l_company == company_name)[0][0]
    return index

# example
index = find_company_index("AAPL.O")
print(index)
```

```python
# calling a set of data specific to a company (example)
# use an array so that it's easier to use loops
df_RIC.get_group(l_company[0])
```

```python
df.to_csv("Data/cleaned_data.csv")
```

```python
df['sigma_t'] = np.sqrt(0.5 * np.square(np.log(df['High']) - np.log(df['Low'])) 
                        - (2 * np.log(2) - 1) * np.square(np.log(df['Close']) - np.log(df['Open'])))
df.head()
```

```python
df_sigma = df.groupby(['RIC', 'Date'])['sigma_t'].mean().reset_index()
df_sigma['ewma'] = 0

for ticker in l_company:
    df_sigma.loc[df_sigma['RIC'] == ticker, 'ewma'] = pd.DataFrame.ewm(df_sigma.loc[df_sigma['RIC'] == ticker, 'sigma_t'], span=20).mean()

df_sigma.head()
```

```python
df_sigma.to_csv("Data/daily_avg_vol.csv")
```
